```python
import data
import panel as pn

pn.extension("plotly")
```

```python
gp = data.GetPlot()

eth_trans = gp.get_eth_trans()

#not loading
top_10 = gp.get_top_10()
btc_eth_price = gp.get_btc_eth_price()
#----------------------
eth_price = gp.get_eth_price()
eth_trans_price = gp.get_eth_trans_price()
eth_price_other_assets = gp.get_eth_price_other_assets()
mining_difficulty = gp.get_mining_difficulty()
mining_to_eth = gp.get_mining_to_eth_close()
heat_map = gp.get_heat_map_pct()

#interfering
box_plot = gp.get_box_plot()
rolling = gp.get_rolling_std()
#-----------------------

sharpe = gp.get_sharpe()
drawdown = gp.get_drawdown()
```

```python
row1 = pn.Row(eth_trans, eth_price)
row2 = pn.Row(eth_price_other_assets,btc_eth_price)
row3 = pn.Row(mining_difficulty,mining_to_eth)
row4 = pn.Row(heat_map,box_plot)
row5 = pn.Row(rolling,sharpe)
row6 = pn.Row(drawdown)
```

```python
column1 = pn.Column(
         '# ETH Trans and Price',
         '## Title 2',
          row1,
          row2
                  
)
```

```python
column2 = pn.Column(
    '# More Price Stuff',
    row3,
    row4
)
```

```python
column3 = pn.Column(
    '#Other neat stuff',
    row5,
    row6
)
```

```python
title = pn.pane.Markdown("""
        ## A small view into the world of Ethereum ("ETH")
        """, 
        width = 800,
)
```

```python
dashboard = pn.Tabs(
        ("Price and Trans", column1),
        ("Title Again", column2),
        ("Yep, another title", [rolling,sharpe,drawdown])
        #("Last one", row2)

)
```

```python
dashboard.show()
```
